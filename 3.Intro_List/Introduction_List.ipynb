## Copy List


```python
mylist =['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
```

```python
mylist1 = mylist #create a new reference "mylist1"
```

```python
id(mylist),id(mylist1) #  the address of both mylist & mylist1 is same
```

```python
mylist2 = mylist.copy() #create a copy of the list
```

```python
id(mylist2) #the add of mylist2 will be different from mylist bcoz
```

- **A shallow copy** means that mylist2 is a new list object with its own memory address, but it contains references to the same elements as mylist.

## All / Any
The **all()** methodreturns:
- **True**- if all element in a list are true     
- **False** - if any element in a list is false      
The **any()** fun returns True if any element in the list is True.If not,any() returns False

```python
L1 = [1,2,3,4,0]
```

```python
all(L1) #return false as one value is false(value 0)
```

```python
any(L1) #return True as we have items in the list with True value
```

```python
L2 = [1,2,3,4,True,False]
```

```python
all(L2) # returns false as one value is false
```

```python
any(L2) # return True as we have items in the list with True value
```

```python
L3 =[1,2,3,True] 
```

```python
all(L3) #return True as all items in the list are True
```

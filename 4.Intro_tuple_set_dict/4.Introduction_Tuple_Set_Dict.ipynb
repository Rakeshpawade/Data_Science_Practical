# Tuple
- 1. Tuple is similar to List except that the objects in tuple are immutable which means we cannot change the elements of a tuple once assigned.
- 2. When we do not want to change the data over time, tuple is a   preferred data type.
- 3. Iterating over the elements of a tuple is faster compared to iterating over a list.

## Set Operation

### Union

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
C = {8,9,10}
```

```python
A|B # union-all element from both sets(no duplicate)
```

```python
A.union(B)
```

```python
A.union(B,C)
```

```python
"""
Update the set calling the update() method with union of A,B & C.
for below example Set A will be updated with union of A,B &C.
"""
A.update(B,C)
A
```

### Intersection

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A & B #intersection-common item from both
```

```python
A.intersection(B) 
```

```python
"""
updates the set calling the intersection_update() method with the intersection A,B
For below example Set A will be updated with the intersection of A & B.
"""
A.intersection_update(B)
A
```

### Difference

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A-B #set of element that are only in A but not in B
```

```python
A.difference(B)
```

```python
B-A #set of element that are only in B but not in A
```

```python
B.difference(A)
```

```python
"""
Updates the set calling the difference_update() method with the difference of set A,B
For below example Set B will be updated with the difference of B & A.
"""
B.difference_update(A)
B
```

### Symmetric Difference

```python
A = {1,2,3,4,5}
B = {4,5,6,7,8}
```

```python
A^B #symmatric difference (set of elements in A and B but not in both)
```

```python
A.symmetric_difference(B)
```

```python
"""
Updates the set calling the symmetric_difference_update() method with the A,B
For below example Set A will be updated with the symmetric difference of A & B.
"""
A.symmetric_difference_update(B)
A
```

### Subset,Superset & Disjoint

```python
A = {1,2,3,4,5,6,7,8,9}
B = {3,4,5,6,7,8}
C = {10,20,30,40}
```

```python
B.issubset(A) # set B all element of B are available in set A
```

```python
A.issuperset(B) # all element of B available in A
```

```python
C.isdisjoint(A) # no common element betn C and A
```

```python
B.isdisjoint(A)
```

## Dictionary

-  Dictionary is a mutable data type in Python.
-  A python dictionary is a collection of key and value pairs separated by a colon (:) & enclosed
   in curly braces {}.
-  Keys must be unique in a dictionary, duplicate values are allowed.

- **mydict = {'A':'one','B':'two','C':'three','D':'four'}**

### All / Any

- The all() method returns:
    - True - If all keys of the dictionary are true
    - False - If any key of the dictionary is false
- The any() function returns True if any key of the dictionary is   True. If not, any() returns False.

```python
mydict1
```

```python
all(mydict1) # will return false as one value is false(value 0)
```
